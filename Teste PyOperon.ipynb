```python
# Importar as bibliotecas necessárias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
print('AQUI')
from pyoperon.sklearn import SymbolicRegressor

# Criar dados de exemplo (uma função simples: y = 2*x + 3 com algum ruído)
np.random.seed(0)
X = np.random.rand(100, 1) * 10
y = 2 * X[:, 0] + 3 + np.random.randn(100) * 0.5
df = pd.DataFrame({'x': X[:, 0], 'y': y})
print(df['x'])

# Inicializar o regressor simbólico
model = SymbolicRegressor(
    #allowed_symbols='add,sub,mul,div'
)

# Ajustar o modelo aos dados
model.fit(df[['x']], df['y'])

# Imprimir o melhor modelo encontrado
print("Melhor modelo encontrado:", model)

# Fazer previsões
predictions = model.predict(df[['x']])

# Visualizar os resultados
plt.scatter(X, y, label='Dados Reais')
plt.scatter(X, predictions, color='red', label='Previsões do Pyperon')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Exemplo Básico de Regressão Simbólica com Pyperon no Colab')
plt.show()
```
